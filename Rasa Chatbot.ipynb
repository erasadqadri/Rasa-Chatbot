# Development of Rasa Chatbot:

```python
import os
import rasa
import nest_asyncio

nest_asyncio.apply()

print('Event Loop is Ready')
```

```python
#to create initiale project first:
from rasa.cli.scaffold import create_initial_project
```

```python
#creating working directory and storing initial project there:
project = 'initial_project'
create_initial_project(project)
```

```python
#as initial project created moving into project directory:
os.chdir(project)
print(os.listdir("."))
```

```python
config = "config.yml"
training_files = "data/"
domain = "domain.yml"
output = "models/"
print(config, training_files, domain, output)
```

# Training Model:

(NOTE: Sometime it may not run on first go. So, try to execute below line 3 time and it will work fine. It is bug in rasa.)

```python
model_path = rasa.train(domain, config, [training_files], output)
print(model_path)
```

```python
#To start chat with our model:
from rasa.jupyter import chat
endpoints = 'endpoints.yml'
chat(model_path, endpoints)
```
